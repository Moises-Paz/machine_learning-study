```python
import numpy as np
import pandas as pd
```

```python
#caminho para o dataset london.csv
path = "D:/programacao/my_programas/FIC_ML/dataset/london.csv"
```

```python
london = pd.read_csv(path)
```

```python
london.head()
```

```python
#1.1 Mostre somente os atletas do sexo feminino.
london.query("Sex=='F'")
```

```python
#Pegando apenas as idades dos atletas de boxing
boxAge = london.query("Sport=='Boxing'")
```

```python
#1.2 Mostre a média de idade dos atletas de boxe.
boxAge[['Age']].mean()
```

```python
#Conversão dos pessoas --> |b para KG
london['Weight'] = london['Weight'] / 2.20462
```

```python
#1.3 a) criando a coluna IMC2 com os resultados do calculo IMC
london['IMC2'] = london['Weight']/((london['Height']/100)**2)
```

```python
#Mostrando o resultado da criaçao da coluna IMC2
london.head()
```

```python
#criando a nova coluna SITUACAO
london['SITUACAO'] = 'NaN'
```

```python
london.head()
```

```python
#1.3 b) Criando uma função que checa todas as situações referente ao IMC2
for index, row in london.iterrows():
    if row['Sex'] == 'F':
        if london.loc[index,'IMC2'] < 19:
            london.loc[index,'SITUACAO'] = "Abaixo do peso"
        elif london.loc[index,'IMC2'] >= 19 and london.loc[index,'IMC2'] <= 23.9:
            london.loc[index,'SITUACAO'] = "Normal"
        elif london.loc[index,'IMC2'] >= 24 and london.loc[index,'IMC2'] <= 28.9:
            london.loc[index,'SITUACAO'] = "Obesidade Leve"
        elif london.loc[index,'IMC2'] >= 29 and london.loc[index,'IMC2'] <= 38.9:
            london.loc[index,'SITUACAO'] = "Obesidade Moderada"
        elif london.loc[index,'IMC2'] > 39:
            london.loc[index,'SITUACAO'] = "Obesidade Mórbida"
    else:
        if row['Sex'] == 'M':
            if london.loc[index,'IMC2'] < 20:
                london.loc[index,'SITUACAO'] = "Abaixo do peso"
            elif london.loc[index,'IMC2'] >= 20 and london.loc[index,'IMC2'] <= 24.9:
                london.loc[index,'SITUACAO'] = "Normal"
            elif london.loc[index,'IMC2'] >= 25 and london.loc[index,'IMC2'] <= 29.9:
                london.loc[index,'SITUACAO'] = "Obesidade Leve"
            elif london.loc[index,'IMC2'] >= 30 and london.loc[index,'IMC2'] <= 39.9:
                london.loc[index,'SITUACAO'] = "Obesidade Moderada"
            elif london.loc[index,'IMC2'] > 40:
                london.loc[index,'SITUACAO'] = "Obesidade Mórbida"
```

```python
#Mostrando a coluna criada e preenchida com suas respectivas situações
london.head()
```

```python
#1.4 - Mostre quantos atletas de cada sexo disputaram o esporte judô (judo)
Mjudo = london.query("Sport=='Judo' and Sex=='M'")
Fjudo = london.query("Sport=='Judo' and Sex=='F'")
```

```python
#Mostrando apenas os atletas do  sexo feminino do judô
Mjudo.Total.count()
```

```python
#Mostrando apenas os atleatas do sexo masculino do judo
Fjudo.Total.count()
```
