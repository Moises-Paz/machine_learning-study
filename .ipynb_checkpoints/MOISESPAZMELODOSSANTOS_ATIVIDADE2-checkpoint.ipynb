```python
#bibliotecas
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
#adquirindo a imagem tucano3.png
tucano = cv2.imread('D:/programacao/my_programas/FIC_ML/imagens/tucano3.png')
```

1 - Utilizando a imagem "tucano3.png" enviada, realize a conversão para o espaço de cores HSV, e utilizando o código enviado
como exemplo, crie uma máscara que destaque SOMENTE a cor AMARELA.

```python
tucanoHSV = cv2.cvtColor(tucano, cv2.COLOR_BGR2HSV)
claro = np.array([13, 175, 0])
escuro = np.array([25, 255, 255])
mask = cv2.inRange(tucanoHSV, claro, escuro)
output = cv2.bitwise_and(tucano, tucano, mask=mask)
cv2.imshow('Cor detectada', np.hstack((tucano, output)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

2 - Utilizando os métodos de redução das dimensões e interpolação explicados, utilize a mesma imagem do exercício 1 para:

```python
#Transformando a imagem para o padrão RGB
tucanoRGB = cv2.cvtColor(tucano, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(tucanoRGB)
```

```python
#Retução da imagem em 15%
tucano_15_AREA = cv2.resize(tucanoRGB, None, fx = .15, fy = .15, interpolation = cv2.INTER_AREA)
tucano_15_AREA = cv2.cvtColor(tucano_15_AREA, cv2.COLOR_BGR2RGB)

tucano_15_CUBIC = cv2.resize(tucanoRGB, None, fx = .15, fy = .15, interpolation = cv2.INTER_CUBIC)
tucano_15_CUBIC = cv2.cvtColor(tucano_15_CUBIC, cv2.COLOR_BGR2RGB)

tucano_15_LINEAR = cv2.resize(tucanoRGB, None, fx = .15, fy = .15, interpolation = cv2.INTER_LINEAR)
tucano_15_LINEAR = cv2.cvtColor(tucano_15_LINEAR, cv2.COLOR_BGR2RGB)

#Salvando cada imagem da redução de 15%
cv2.imwrite("tucano_15_AREA.png", tucano_15_AREA)
cv2.imwrite("tucano_15_CUBIC.png", tucano_15_CUBIC)
cv2.imwrite("tucano_15_LINEAR.png", tucano_15_LINEAR)

#-----------------------------------------------------------------------------------------------------------------------------
#Retução da imagem em 30%
tucano_30_AREA = cv2.resize(tucanoRGB, None, fx = .30, fy = .30, interpolation = cv2.INTER_AREA)
tucano_30_AREA = cv2.cvtColor(tucano_30_AREA, cv2.COLOR_BGR2RGB)

tucano_30_CUBIC = cv2.resize(tucanoRGB, None, fx = .30, fy = .30, interpolation = cv2.INTER_CUBIC)
tucano_30_CUBIC = cv2.cvtColor(tucano_30_CUBIC, cv2.COLOR_BGR2RGB)

tucano_30_LINEAR = cv2.resize(tucanoRGB, None, fx = .30, fy = .30, interpolation = cv2.INTER_LINEAR)
tucano_30_LINEAR = cv2.cvtColor(tucano_30_LINEAR, cv2.COLOR_BGR2RGB)

cv2.imwrite("tucano_30_AREA.png", tucano_30_AREA)
cv2.imwrite("tucano_30_CUBIC.png", tucano_30_CUBIC)
cv2.imwrite("tucano_30_LINEAR.png", tucano_30_LINEAR)
```

```python
def plot_imagens(img1, tit1, img2, tit2, img3, tit3, img4, tit4, tamanho_fig, ticks = False):
 plt.figure(figsize = tamanho_fig)

 img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
 img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
 img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
    
 plt.subplot(1, 4, 1), plt.imshow(img1), plt.title(tit1)
 if not ticks:
     plt.xticks([]), plt.yticks([])
 plt.subplot(1, 4, 2), plt.imshow(img2), plt.title(tit2)
 if not ticks:
     plt.xticks([]), plt.yticks([])
 plt.subplot(1, 4, 3), plt.imshow(img3), plt.title(tit3)
 if not ticks:
     plt.xticks([]), plt.yticks([])
 plt.subplot(1, 4, 4), plt.imshow(img4), plt.title(tit4)
 if not ticks:
     plt.xticks([]), plt.yticks([])
 plt.show()
```

Plotando Imagens reduzidas em 15%


```python
plot_imagens(tucanoRGB, "Original", tucano_15_AREARGB, "AREA", tucano_15_CUBICRGB, "CUBIC" , tucano_15_LINEARRGB, "LINEAR", (15,15), False)
```
